```python
def extract_(pocket_file_path, pdb_file_path):
    # get list of aa in pocket
    # or get list of aa in complete .ent file(pdb)
    # ATOM ()
    with open(pocket_file_path, "r", encoding="utf8") as fp:
        pocket_lines = [line.strip() for line in fp]
    pocket_list = []
    for line in pocket_lines:
        if line[0:6].strip() != 'ATOM': continue
        atom_name = line[12:16].strip()
        res_name = line[17:20].strip()
        chain = line[21:22].strip()
        res_id = int(line[22:26])
        pocket_list.append((atom_name, res_name, chain, res_id))

    with open(pdb_file_path, "r", encoding="utf8") as fp:
        pdb_lines = [line.strip() for line in fp]
    pdb_list = []
    for line in pdb_lines:
        if line[0:6].strip() != 'ATOM': continue
        atom_name = line[12:16].strip()
        res_name = line[17:20].strip()
        chain = line[21:22].strip()
        res_id = int(line[22:26])
        pdb_list.append((atom_name, res_name, chain, res_id))

    return pocket_list, pdb_list
```

```python
# get pdb file of error pdb_id

import requests
from tqdm import tqdm

process_file = "pdb_id_error.log"
unprocess_pdb_id = []

error_id = []


with open(process_file, 'r', encoding='utf8') as f:
    all_lines = [line.strip() for line in f]

for line in all_lines:
    pdb_id = line.split("ERROR: PDB id")[1][:6].strip()
    unprocess_pdb_id.append(pdb_id)

# print(len(unprocess_pdb_id))
# print(len(set(unprocess_pdb_id)))
unprocess_pdb_id = list(set(unprocess_pdb_id))
for pdb_id in tqdm(unprocess_pdb_id):
    flag = False
    try_times = 0
    while not flag and try_times < 10:
        output_file = f"./pdb_files/pdb{pdb_id}.ent"

        url = f"https://files.rcsb.org/download/{pdb_id}.pdb"

        response = requests.get(url)
        try_times += 1

        if response.status_code == 200:
            flag = True
            with open(output_file, "wb") as f:
                f.write(response.content)
        else:
            if try_times == 10:
                error_id.append(pdb_id)

print(error_id)
        
```

```python
import csv
from io import StringIO

from Bio import SeqIO


def get_chain_fasta(fasta_sequence, chain):
    fasta_list = fasta_sequence.split(">")[1:]
    temp_dict = {}
    for ff in fasta_list:
        ff = ff.strip()
        f_chain = str(ff.split("\n")[0].split(":")[1].strip())
        f_seq = "".join(ff.split("\n")[1:])
        temp_dict[f_chain] = f_seq
    return temp_dict[str(chain)]

def get_start_idx(pdb_file, chain):
    with open(pdb_file, "r", encoding="utf8") as f:
        for line in f:
            line = line.strip()
            if line[0:6].strip() == 'ATOM' and line[21:22].strip() == chain:
                # print(int(line[22:26]))
                return str(line[22:26])
    return None

def get_fasta_from_pdb_biopython_no_file(pdb_id, chain):
    pdb_file = f"./pdb_files/pdb{pdb_id}.ent"
    with open(pdb_file, "r") as file:
        pdb_content = file.read()

    pdb_handle = StringIO(pdb_content)
    fasta_handle = StringIO()
    SeqIO.convert(pdb_handle, "pdb-atom", fasta_handle, "fasta")

    fasta_sequence = fasta_handle.getvalue()

    if fasta_sequence:
        start = get_start_idx(pdb_file, chain)
        fasta = get_chain_fasta(fasta_sequence, chain)
        if start:
            return fasta, start
        else:
            return "", -100
    else:
        return "", -100

process_file = "pdb_id_error.log"
unprocess_pdb_id_chain = []

with open(process_file, 'r', encoding='utf8') as f:
    all_lines = [line.strip() for line in f]

csv_name = "crossdocked2fasta_pocket10.csv" 
with open(csv_name, 'a', newline='') as csvfile:
    writer = csv.writer(csvfile)
    for line in all_lines:
        pdb_id = line.split("ERROR: PDB id")[1][:6].strip()
        chain = line.split("_")[1].strip()
        if pdb_id == "4abz":
            continue
        fasta, start = get_fasta_from_pdb_biopython_no_file(pdb_id, chain)
        writer.writerow([pdb_id, chain, str(fasta), int(start)])
```

```python
import csv
import os
import sys

from tqdm import tqdm

exist_pdb, exist_pdb_chain = [], []

crossdocked_dir = "../crossdocked_pocket10/"
subdir_list = os.listdir(crossdocked_dir)

for subdir in tqdm(subdir_list):
    temp_path = os.path.join(crossdocked_dir, subdir)
    if os.path.isfile(temp_path): continue

    file_list = os.listdir(temp_path)
    for file_name in file_list:
        if file_name.endswith(".sdf"): continue
        protein_filename = file_name
        pdb_id = str(protein_filename.split("_")[0].strip())
        chain = str(protein_filename.split("_")[1].strip())
        if (pdb_id, chain) not in exist_pdb_chain:
            exist_pdb_chain.append((pdb_id, chain))
        if pdb_id not in exist_pdb:
            exist_pdb.append(pdb_id)

with open("pdb_id_chain.txt", "w", encoding="utf8") as fw:
    for ee in exist_pdb_chain:
        print(ee, file=fw)

with open("pdb_id.txt", "w", encoding="utf8") as fw:
    for ee in exist_pdb:
        print(ee, file=fw)
```

```python
import csv

def count_status_occurrences(csv_file_path):
    status_counts = {}
    with open(csv_file_path, newline='') as csvfile:
        reader = csv.DictReader(csvfile)
        for row in reader:
            status = row['status']
            status_counts[status] = status_counts.get(status, 0) + 1

    return status_counts

csv_file_path = "./crossdocked_data_status.csv"
status_counts_dict = count_status_occurrences(csv_file_path)

print(status_counts_dict)
```
